#### 10. How can you tell the difference between break and continue?

'break' and 'Continue' are control flow statment that alter the execution of loops. 
1. Break: 
    - The 'Break' statment is used to exit or terminate the current loop.
      When a break statment is encountered within a loop,the loop is immediately terminated,and the program continues execution from the next statment following the loop.
      It is commonly used to stop the loop when a specific condition is met,or when further iteration are unnecessary.
    
Example:
    
    for i in range(1,6):
        if i == 3:
            break
    Print(i)

2. Continue:
    -The 'Continue' statment is used to skip the current iteration of the loop and move to the next iteration.
     When a 'Continue' statement is encountered within a loop,the remaining statments within the loop for that particular iteration are skipped,and the loop proceeds to the next iteration.
     It is commonly used to skip certain iteration based on specific condition,without exiting the loop entirely.
    
Example:
     
        For i in the range(1,6):
            if i == 3
            continue
        print(i) 

#### 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

1. range(10):
    - This expression represents  range from 0 but not including, the specified end value of 10.
    - It increments by 1 for each iteration
    - The start value of 0 and the step value of 1 are implied and not explicitly mentioned.

```python
for i in range(10):
    print(i)              #This loop will iterate from 0 to 9 printing the values 0 to 9.
```

2. range(0,10):
    -This expression explictily specified bothe the start and end value of the range.
    -It starts from 0 and goes up to,but not including,the specified and end value of 10.
    -It increments by 1 for each iteration.
    
    Example

```python
for i in range (0,10):
    print (i)
```

3. range(0,10,1):
    -This expression explicitly specifies the start value,end value,and step value of the range.
    -It starts from 0 goes upto,but not inckuding, the specified end value of 10,and increments by 1 for each iteration.
    
    Example:

```python
for i in range(0,10,1):
    print(i)
```

While the expressions 'range(10)','range(0,10)','range(0,10,1)' may have slightly diffrent ways of specifyng the range,they produce the same sequence of valye and can be used interchangeable in for a 'for'loop. The default start value is 0,the default step value is 1,and the end value id not inclusive.
